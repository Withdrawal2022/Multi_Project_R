```python
import pandas as pd
data = pd.read_csv('data/2023 QS World University Rankings.csv')

first_two_columns = data.iloc[:240,:2]
other_we_want = data.iloc[:240,4:15:2]
last_column = data.iloc[:240,-1]
data_used = pd.concat([last_column, first_two_columns,other_we_want],axis=1)

data_used.head()
```

```python
first_two_columns = data.iloc[:240,:2]
other_we_want = data.iloc[:240,4:20:2]
last_column = data.iloc[:240,-1]
data_used = pd.concat([last_column, first_two_columns,other_we_want],axis=1)

data_used.head()
```

```python
import numpy as np
from sklearn.preprocessing import StandardScaler
standardized = StandardScaler().fit_transform(data_used.iloc[:,3:])

standardized_df = pd.DataFrame(standardized,columns=data_used.columns[3:])
standardized_df.head()
```

# PCA

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=4)
result_skPCA = pca.fit_transform(standardized_df)
result_skPCA.shape
```

```python
data_used["PCA_Scores"]=result_skPCA@(pca.explained_variance_ratio_/np.sum(pca.explained_variance_ratio_)).T
```

```python
sorted_data_used = data_used.sort_values("PCA_Scores",ascending=False)
```

# Desicion Tree

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
```

```python
X = data_used.iloc[:,3:11]
X = np.array(X)
```

```python
y = np.array(data_used["Rank"])
```

```python
# 创建决策树分类器对象
clf = DecisionTreeClassifier()

# 使用训练数据拟合分类器模型
clf.fit(X, y)
```

```python
# 获取特征重要性
feature_importance = clf.feature_importances_
importances = []
# 打印各个特征的重要性
for i, importance in enumerate(feature_importance):
    print(f"特征 {i+1}: {importance}")
    importances.append(importance)
```

```python
data_used["DT_Scores"] = np.dot(data_used.iloc[:, 3:11], importances)
```

```python
sorted_data_used = data_used.sort_values("DT_Scores",ascending=False)
```

```python
sorted_data_used.head(10)
```
